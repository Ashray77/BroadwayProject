# Game Recommendation System (Content-Based)

```python
import pandas as pd
import numpy as np
df = pd.read_csv('steam_games.csv')
```

```python
df.head(10)
```

```python
df['game_description'].head(5)
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')
```

```python
#Replace NaN with an empty string
df['game_description'] = df['game_description'].fillna('')
```

```python
#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['game_description'])
```

```python
#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
#Construct a reverse map of indices and games names
indices = pd.Series(df.index, index=df['name']).drop_duplicates()
```

```python
# Function that takes in game names as input and outputs most similar games
def game_recommendations(name, cosine_sim=cosine_sim):
    # Get the index of the game that matches the title
    idx = indices[name]

    # Get the pairwsie similarity scores of all games with that game
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the games based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # Get the game indices
    game_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar games
    return df['name'].iloc[game_indices]
```

```python
game_recommendations('DOOM')
```

```python
game_recommendations('GOD EATER 3')
```

```python
game_recommendations('Call of Duty®: Modern Warfare® Remastered')
```

```python
game_recommendations('For The King')
```
